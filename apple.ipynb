```python
import pandas as pd
import seaborn as sns
```

```python
cf=pd.read_csv('apple.csv')
```

```python
cf.columns
```

```python
top_5_star_rating=cf.nlargest(5,'Star Rating')[['Product Name','Star Rating']]
top_5_star_rating
```

```python
top_5_star_rating=cf.nsmallest(5,'Star Rating')[['Product Name','Star Rating']]
top_5_star_rating
```

```python
top_5_star_rating=cf.nlargest(5,'Star Rating')[['Product Name','Star Rating','Number Of Ratings']]
top_5_star_rating
```

```python
top_5_star_rating=cf.nsmallest(5,'Star Rating')[['Product Name','Star Rating','Number Of Ratings']]
top_5_star_rating
```

```python
highest_review_phone=cf[cf["Number Of Reviews"]==cf["Number Of Reviews"].max()]["Product Name"]
highest_review_phone
```

```python
lowest_review_phone=cf[cf["Number Of Reviews"]==cf["Number Of Reviews"].min()]["Product Name"]
lowest_review_phone
```

```python
relation_price_rating=cf[['Sale Price','Number Of Ratings']].corr()
sns.heatmap(relation_price_rating,annot=True,cmap="coolwarm")
relation_price_rating
```

```python
relation_price_rating=cf[['Sale Price','Number Of Reviews']].corr()
sns.heatmap(relation_price_rating,annot=True,cmap="coolwarm")
relation_price_rating
```

```python
relation_price_reviews=cf[['Sale Price','Number Of Reviews']].corr()
sns.heatmap(relation_price_reviews,annot=True,cmap="coolwarm")
relation_price_reviews
```

```python
relation_discount_reviews=cf[['Sale Price','Number Of Reviews']].corr()
sns.heatmap(relation_discount_reviews,annot=True,cmap="coolwarm")
relation_discount_reviews
```
